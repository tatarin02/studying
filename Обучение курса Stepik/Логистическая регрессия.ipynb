Почему сигмойда

σ(x) = 1 / (1 + e^(-z))

z=w⊤x+b


### Построение графика сигмойдной функции

```python
# Создание z и σ для графика

import numpy as np

z = np.linspace(-10, 10, 1000)
sigm = np.array([1/(1 + np.exp(-z)) for z in z])
```

```python
# Создание графика
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(6, 4), dpi=120)

ax.plot(z, sigm, label='sigmoid func')
ax.legend()
```

```python
plt.close(fig)
```
