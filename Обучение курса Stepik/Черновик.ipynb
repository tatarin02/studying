```python
n = int(input())
low = 10
high = 90

step = 100 / n

low_indxs = []
high_indxs = []

for i in range(1, n+1):
    if step * i < low:
        low_indxs.append(i-1)
    elif step * i > high:
        high_indxs.append(i-1)
print(low_indxs, high_indxs)
```

```python

import pandas as pd
from sklearn.preprocessing import OrdinalEncoder

X = pd.DataFrame({"size": ["S", "M", "XL", "L", "M"]})
enc = OrdinalEncoder(categories=[["XS", "S", "M", "L", "XL"]],
                     handle_unknown="use_encoded_value",
                     unknown_value=-1)
X_enc = enc.fit_transform(X)

X_enc
```
