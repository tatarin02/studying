```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "age": [25, np.nan, 41, 39, np.nan],
    "income": [50_000, 62_000, np.nan, 70_000, 65_000],
    "city": ["A", "A", "B", np.nan, "B"]
})

display(df)

# Среднее для income
df["income"] = df["income"].fillna(df["income"].mean())

# Медиана для age
df["age"] = df["age"].fillna(df["age"].median())

# Мода для city (если несколько мод — берем первую)
df["city"] = df["city"].fillna(df["city"].mode().iloc[0])

# Групповая медиана age по city
# df["age"] = df["age"].fillna(
    # df.groupby("city")["age"].transform("median")
# )

display(df)
```
